```python
import matplotlib.pyplot as plt 
import numpy as np 

from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.sans-serif'] = ['Times New Roman']
```

```python
cmaps = ['bone','Spectral','autumn','rainbow']
```

```python

plt.figure(1, figsize=(20,10))

cmap = plt.get_cmap(cmaps[0])


colors = [cmap(i) for i in np.linspace(0, .5, 4)]

counts = [.29,.24,.14,.33]

labels = ["\n\n\n Effectively \n Single","\n\nMerge","\n\n\n\n Accretion & \n Spin Up \n or CE","\n\n\n Envelope \n Stripping"]

w,l,p = plt.pie(counts, labels=labels, autopct='%1.1f%%', shadow=True, colors=colors,explode=[.05,.05,.05,.1],pctdistance=0.5,labeldistance=0.5,textprops={'fontsize':16,'color':'white'});


p[3].set_size(20)
l[3].set_size(20)
p[3].set_position([0.2,-0.45])
l[3].set_position([0.2,-0.45])

plt.savefig('SanaPie.png')
```
